```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 데이터 출처
# https://www.kaggle.com/datasets/alfredkondoro/seoul-historical-weather-data-2024
weather_file_16_18 = "./data/seoul 2016-01-01 to 2018-01-01.csv"

def import_weather(address:str) -> pd.DataFrame:
    weather_data = pd.read_csv(address)

    weather_data.loc[:, "year"] = weather_data["datetime"].str[:4].astype(int)
    weather_data.loc[:, "month"] = weather_data["datetime"].str[5:7].astype(int)
    weather_data.loc[:, "date"] = weather_data["datetime"].str[8:].astype(int)

    front = ["year", "month", "date"]
    others = [c for c in weather_data.columns if c not in front]
    weather_data = weather_data[front + others]

    weather_data = weather_data.drop(["datetime"], axis=1)

    return weather_data

weather_data_16_18 = import_weather(weather_file_16_18)
weather_data_16_18.head()
```

```python
weather_file_18_20 = "./data/seoul 2018-01-01 to 2020-01-01.csv"

def import_weather(address:str) -> pd.DataFrame:
    weather_data = pd.read_csv(address)

    weather_data.loc[:, "year"] = weather_data["datetime"].str[:4].astype(int)
    weather_data.loc[:, "month"] = weather_data["datetime"].str[5:7].astype(int)
    weather_data.loc[:, "date"] = weather_data["datetime"].str[8:].astype(int)

    front = ["year", "month", "date"]
    others = [c for c in weather_data.columns if c not in front]
    weather_data = weather_data[front + others]

    weather_data = weather_data.drop(["datetime"], axis=1)

    return weather_data

weather_data_18_20 = import_weather(weather_file_18_20)
weather_data_18_20.head()
```

```python
weather_df = pd.concat([weather_data_16_18, weather_data_18_20], axis=0)
weather_df['year'].value_counts()
```

```python
weather_df = weather_df[weather_df['year'] != 2016]
weather_df = weather_df[weather_df['year'] != 2020]
```

```python
weather_df[
    (weather_df['year'] == 2018) &
    (weather_df['month'] == 1) &
    (weather_df['date'] == 1)
]
```

```python
weather_df = weather_df.drop_duplicates(
    subset=['year', 'month', 'date'],
    keep='first'   # 첫 번째 남기고 나머지 삭제
)
```
